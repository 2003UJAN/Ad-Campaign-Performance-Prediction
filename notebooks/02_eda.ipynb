```
# --------------------------------------------
# 02_eda.ipynb
# --------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("ads_synthetic.csv")
df.head()

# --------------------------
# 1. Basic Info
# --------------------------
print(df.info())
print(df.describe())

# --------------------------
# 2. Distribution Plots
# --------------------------
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Performance")
plt.title("Ad Performance Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df["Budget"], bins=30, kde=True)
plt.title("Ad Budget Distribution")
plt.show()

# --------------------------
# 3. CTR vs Conversions
# --------------------------
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="CTR", y="Conversions", hue="Performance")
plt.title("CTR vs Conversions")
plt.show()

# --------------------------
# 4. Ad Type vs Performance
# --------------------------
plt.figure(figsize=(7,5))
sns.countplot(data=df, x="AdType", hue="Performance")
plt.title("AdType vs Performance")
plt.show()

# --------------------------
# 5. City-based performance
# --------------------------
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x="City", y="Impressions")
plt.title("City-wise Ad Impressions")
plt.xticks(rotation=45)
plt.show()
```
